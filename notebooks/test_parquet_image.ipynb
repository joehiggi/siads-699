# Import Required Libraries
Import necessary libraries such as pandas for handling Parquet files, and PIL for image processing.

```python
# Import Required Libraries
import pandas as pd
from PIL import Image
import io
```

# Load the Parquet File
Use pandas to read the Parquet file into a DataFrame.

```python
# Load the Parquet File
df = pd.read_parquet("data/raw/train-00000-of-00005.parquet")
print(f"Total rows: {len(df)}")
print(f"Columns: {list(df.columns)}")
```

# Extract Image Data
Extract the image data from the DataFrame, assuming it's stored as bytes in the 'image' column.

```python
# Extract Image Data
first_image_data = df["image"].iloc[0]
if isinstance(first_image_data, dict) and "bytes" in first_image_data:
    image_bytes = first_image_data["bytes"]
else:
    image_bytes = first_image_data
print(f"Image size: {len(image_bytes)} bytes")
```

# Display the Image
Use PIL to display the extracted image.

```python
# Display the First 5 Images
from IPython.display import display

for i in range(5):
    first_image_data = df["image"].iloc[i]
    if isinstance(first_image_data, dict) and "bytes" in first_image_data:
        image_bytes = first_image_data["bytes"]
    else:
        image_bytes = first_image_data
    image = Image.open(io.BytesIO(image_bytes))
    print(f"Image {i+1}:")
    display(image)
```
